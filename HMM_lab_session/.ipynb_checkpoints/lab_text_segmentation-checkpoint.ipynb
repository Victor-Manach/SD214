## Q1
We have that the decoding necessarily begins in the state 1 (and each e-mail contains a header), therefore we have $$\pi=\begin{pmatrix}1&0\end{pmatrix}$$

## Q2
The probability to move from state $1$ to state $2$ is given by $A(1,2)=0.000781921964187974$. This probability is very small because it corresponds to the probability to change state between two characters in the e-mail. There is only one transition from the header to the body, therefore the transition is quite rare. Once the transition has been made and that we are in the state $2$, it is not possible to go back to the header, thus the transition $2\rightarrow1$ is not possible and we stay in the state $2$ with probability $1$.

## Q3
There are N different characters and two different states (state 1 and 2), so the size of the matrix B has N rows and 2 columns. $$\text{B is size}\ (N,2).$$

## Viterbi Algorithm

```python
def viterbi(X,Pi0,A,P):
    """
        Viterbi Algorithm Implementation

        Keyword arguments:
            - obs: sequence of observation
            - states:list of states
            - start_prob:vector of the initial probabilities
            - trans: transition matrix
            - emission_prob: emission probability matrix
        Returns:
            - seq: sequence of state
    """

    # pour eviter d avoir des valeurs nulles dans le log
    realmin = np.finfo(np.double).tiny
    A = np.log(A+realmin)
    Pi0 = np.log(Pi0+realmin)
    P = np.log(P+realmin)
    taille = np.shape(X) #X.shape[0]
    T = taille[0] # nombre d observations
    N = Pi0.shape[0] # nombre des etats du modele
    # print('(%s,%s)'%(T,N))
    # Initialisation
    logl = np.zeros((T,N))
    bcktr = np.zeros((T-1,N))
    
    logl[0,:]=Pi0+P[X[0],:]
    for t in range(1,T):
        tmp = A + logl[t-1,:].reshape((N,1))
        logl[t,:] = P[X[t],:] + np.amax(tmp, axis=0)
        bcktr[t-1,:] = np.argmax(tmp, axis=0)
        
    path = np.zeros(T, dtype=int)
    path[T-1] = np.argmax(logl[T-1,:])
    for t in range(T-2,-1,-1):
        path[t]=bcktr[t,path[t+1]]
    path+=1 # to have states 1 and 2 instead of 0 and 1
    return logl, path
```

## Q5
To segment the mail in more parts, I would create a state for each part (3 states if we consider header, body, signature). In this case, the transition matrix $A$ would of size $(3,3)$ and $B$ of size $(N,3)$ with N being the number of different characters. 

The matrix $A$ would be of the following form: $$A\approx \begin{pmatrix}1-\epsilon_1 &\epsilon_1 & 0 \\ 0 &1-\epsilon_2 & \epsilon_2 \\ 0 & 0 & 1\end{pmatrix}$$ 

Indeed, one cannot return to a previous state, and the probabilities of changing states are quite small (as the probability to move from the header to the body was small, we would have 2 small probabilities for the 2 transitions). Thus, $\epsilon_i$ is a small positive integer (we had $\epsilon\approx0.00078$ for the previous part). 
In addition, one cannot move from the header to the signature directly (we suppose that there is always a body). 

Finally, we also consider that each mail contains a header, so the vector of initial probabilities is $$\pi=\begin{pmatrix}1&0&0\end{pmatrix}$$

The probability of observations for each state (matrix $B$) should be estimated the same way it has been done for the previous part (on a small number of mails).

## Q6
I would add another state for the included mail. With the previous example, we would have 4 states: header, included mail, body and signature. Thus, the transition matrix would have the following form: $$A=\begin{pmatrix}a_{11}&a_{12}&a_{13}&0\\0&a_{22}&a_{23}&a_{24}\\0&a_{32}&a_{33}&a_{34}\\0&0&0&1\end{pmatrix}$$

Indeed, it is possible to move from the header to the mail included or directly to the body, and there is a transition from the body to the mail included and vice-versa. When we reach the signature state, it is not possible to go back to previous states.

In order to get a more accurate cut, we could use bigram to get more context around the character, which would be especially usefull for the '>' character for the mail included state.

Finally, we have seen previously that most separation are made between lines and not in the middle of a line. Therefore, we could look at entire lines instead of looking at single characters. In this case, all lines starting with '>' would be included mail, and it would solve the issue of wrong separation that cut the middle of a line.

## Bonus: Unsupervised learning

### Alpha calculation

```python
from scipy.special import logsumexp

def alpha_calculation(X,Pi0,A,P):
    realmin = np.finfo(np.double).tiny
    A = np.log(A+realmin)
    Pi0 = np.log(Pi0+realmin)
    P = np.log(P+realmin)
    T = X.shape[0] # nombre d observations
    N = Pi0.shape[0] # nombre des etats du modele
    
    # Initialisation
    alphas = np.zeros((T,N))
    
    alphas[0,:] = Pi0 + P[X[0],:]
    for t in range(1,T):
        alphas[t,:] = P[X[t],:] + logsumexp(A+alphas[t-1,:].reshape((1,N)), axis=1)
        
    return alphas
```

```python
X = mails[0]
alphas = alpha_calculation(X,Pi0,A,P)
```

### Beta calculation

```python
def beta_calculation(X,A,P):
    realmin = np.finfo(np.double).tiny
    A = np.log(A+realmin)
    P = np.log(P+realmin)
    T = X.shape[0] # nombre d observations
    N = Pi0.shape[0] # nombre des etats du modele
    
    # Initialisation
    betas = np.zeros((T,N))
    
    betas[T-1] = np.ones((1,N))
    for t in range(T-2,-1,-1):
        betas[t,:] = logsumexp(A+P[X[t+1],:].reshape((1,N))+betas[t+1,:].reshape((1,N)),axis=1)
    return betas
```

```python
betas = beta_calculation(X,A,P)
```

### Reestimation of the coefficients

```python
def update(alphas, betas):
    T,N = alphas.shape
    gammas = np.zeros((T,N))
    gammas = alphas+betas-logsumexp(alphas+betas, axis=1, keepdims=True)
    return gammas
```

```python
gammas = update(alphas, betas)
```
